Assumptions for Linear Regression

Linearity Linear regression needs the relationship between the independent and dependent variables to be linear. Let's use a pair plot to check the relation of independent variables with the annual variable

```python
# visualize the relationship between the features and the response using scatterplots
p = sns.pairplot(df, x_vars=[Jan-Feb','Mar-May','Jun-Sep','Oct-Dec'], y_vars='ANNUAL', size=10, aspect=0.7)
```

By looking at the plots we can see that with the annual variable the none of the independent variables form an accurately linear shape but june - september do still better than rest all which seems to hardly have any specific shape. So it shows that a linear regression fitting might not be the best model for it. A linear model might not be able to efficiently explain the data in terms of variability, prediction accuracy etc.

A tip is to remember to always see the plots from where the dependent variable is on the y axis. Though it wouldn't vary the shape much but that's how linear regression's intuition is, to put the dependent variable as y and independents as x(s).


```python
# Fitting the linear model
x = df.drop(["ANNUAL",'STATE_UT_NAME','DISTRICT'],axis=1)
y = df.ANNUAL
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state = 0,test_size=0.25)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X_train,y_train)
y_pred = regr.predict(X_train)
```

```python
print("R squared: {}".format(r2_score(y_true=y_train,y_pred=y_pred)))
```

Mean of Residuals - Residuals as we know are the differences between the true value and the predicted value. One of the assumptions of linear regression is that the mean of the residuals should be zero. So let's find out.

```python
residuals = y_train.values-y_pred
mean_residuals = np.mean(residuals)
print("Mean of Residuals {}".format(mean_residuals))
# Very close to zero so all good here.
```

```python
p = sns.scatterplot(y_pred,residuals)
plt.xlabel('y_pred/predicted values')
plt.ylabel('Residuals')
plt.ylim(-10,10)
plt.xlim(0,26)
p = sns.lineplot([0,26],[0,0],color='blue')
p = plt.title('Residuals vs fitted values plot for homoscedasticity check')
```

```python
from statsmodels.stats import diagnostic as diag
min(diag.acorr_ljungbox(residuals , lags = 40)[1])
#Since p value is not less than 0.05 we do not reject the null hypothesis that error terms are autocorrelated.
```
